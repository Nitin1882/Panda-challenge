```python
# Age Demographics

# Establish bins for ages
# Categorize the existing players using the age bins. Hint: use pd.cut()
# Calculate the numbers and percentages by age group
# Create a summary data frame to hold the results
# Optional: round the percentage column to two decimal points
# Display Age Demographics Table
bins = [0, 10, 14, 19, 24, 29,34,39,50]
labels = ['<10', '10-14', '15-19', '20-24', '25-29', '30-34', '35-39', '40+']

sn_df = purchase_data.loc[:,['SN', 'Age']]

sn_age = {}

for x in range(len(sn_df)):
    
    sn = sn_df.iloc[x,0]
    age = sn_df.iloc[x,1]
    
    if sn not in sn_age:
        sn_age[sn] = age
        
len(sn_age)


names = []

ages = []

for x in sn_age:
    
    names.append(x)
    ages.append(sn_age[x])
    
# Answer: Age groups (unique)    
names_age_df = pd.DataFrame({"SN": names, "Age" : ages})

names_age_df["Age Group"] = pd.cut(names_age_df['Age'], bins, labels = labels)

names_age_df.head(576)

agegroup = names_age_df.groupby(['Age Group'])

agegroup.count().head(10)

agegroup_df = pd.DataFrame(agegroup.count())


# Answer: Age Group (unique) Percentages
agegroup_df['Percents'] = agegroup_df['Age']/len(sn_age) * 100

del agegroup_df["SN"]

agegroup_df = agegroup_df.rename(columns = {'Age' : 'Total Count', 'Percents': 'Percentage of Players'})

agegroup_df["Percentage of Players"] = agegroup_df["Percentage of Players"].map("{:.2f}%".format)

agegroup_df

# purchase_data['Age_Group'] = pd.cut(purchase_data['Age'], bins=bins, labels=names)
# print(purchase_data)
#purchase_data['binned'] = np.searchsorted(bins, purchase_data['Age'].values)
#print (purchase_data)
# testdata=purchase_data.groupby(['Age_Seg'])
```

```python
# Seems like groupby but for SN
# You need to be able to sort through a column that accounts for all the values given not just count
# or average or total but all three at the same time

dfcopy = purchase_data.copy()

sngb = dfcopy.groupby(['SN'])

# This will be a column of the final dataframe
sngb[['Purchase ID']].count().head()

# This will be a column of the final dataframe
sngb[['Price']].mean().head()

# This will be a column of the final dataframe
sngb[['Price']].sum().head()

df_sorted = pd.DataFrame()

df_sorted['Purchase Count'] = sngb['Purchase ID'].count()

df_sorted['Average Purchase Price'] = sngb['Price'].mean()

df_sorted['Total Purchase Value'] = sngb['Price'].sum()

df_sorted.head()

# Answer: Sorted by Purchase Count, Average Purchase Price, and Total Purchase Value
df_finalsorted = df_sorted.sort_values(['Total Purchase Value']
                                      , ascending = False)

df_finalsorted.head()
```

```python
# Sort purchase count column in descending order

item_idnameprice = purchase_data.loc[:, ['Item ID','Item Name', 'Price']]

item_idnameprice

items_groupby = item_idnameprice.groupby(['Item ID', 'Item Name'])

# Answer: Item count, to be sorted
items_groupby.count().head(10)

# Answer: Item price, to be sorted
items_groupby.mean().head(10)

# Answer: Total Item Price, to be sorted
items_groupby.sum().head(10)

df_mostpop = pd.DataFrame(items_groupby.count())

df_mostpop = df_mostpop.rename(columns = {'Price': 'Purchase Count'})

df_mostpop['Item Price'] = items_groupby.mean()

df_mostpop['Total Purchase Value'] = items_groupby.sum()

df_mostpop = df_mostpop.sort_values('Purchase Count', ascending = False)

df_mostpop.head()
```
